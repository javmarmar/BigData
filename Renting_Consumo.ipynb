```python
import pandas as pd
arval_table = pd.read_csv('fuel_joined_toJavi.txt', sep='|',engine= 'python')

arval_table.head()
```

```python
consumos=arval_table.loc[:,['DateHeure','IdVehicule','KmSaisie','Quantite']]
consumos.head()
```

```python
consumos.dtypes
```

```python
import datetime
consumos['Fecha'] = consumos['DateHeure'].str.split(':').str[0]
consumos['Día'] = consumos['Fecha'].str.slice(0,2)
consumos['Mes'] = consumos['Fecha'].str.slice(2,5)
consumos['Año'] = consumos['Fecha'].str.slice(5,9)
consumos['Hora'] = consumos['DateHeure'].str.split(':').str[1]
consumos['Min'] = consumos['DateHeure'].str.split(':').str[2]
consumos['Seg'] = consumos['DateHeure'].str.split(':').str[3]
consumos.head()
```

```python
consumos['Fecha'] =  consumos['Día'].map(str)+str('/')+consumos['Mes'].map(str)+str('/')+consumos['Año'].map(str)+str('-')+consumos['Hora'].map(str)+str(':') + consumos['Min'].map(str)+str(':') + consumos['Seg'].map(str)
consumos.head()
```

```python
consumos['DateHeure'] =  pd.to_datetime(consumos['Fecha'], infer_datetime_format=True)
consumos.head()
```

```python
consumos.drop(['Fecha','Día','Mes','Año','Hora','Min','Seg'], axis=1, inplace=True)
consumos.head()
```

```python
consumos.sort_values(by='DateHeure', inplace=True)
consumos.head()
```

```python
consumos.groupby('IdVehicule')['KmSaisie'].unique()
```

```python
tabla_km=consumos.loc[:,['DateHeure','IdVehicule','KmSaisie']]
tabla_km.sort_values(by='DateHeure', inplace=True)
tabla_km.head()
```

```python
tabla_km=tabla_km[tabla_km.KmSaisie != 0]
tabla_km.head()
```

```python
import numpy as np
```

```python
def flag_outlier(x):
    lower_limit  = np.mean(x) - np.std(x) * 2 
    upper_limit = np.mean(x) + np.std(x) * 2
    return (x>upper_limit)| (x<lower_limit)
```

```python
tabla_km['Flag'] = tabla_km.groupby('IdVehicule')['KmSaisie'].apply(flag_outlier)
```

```python
tabla_km=tabla_km[tabla_km.Flag==False]
```

```python
tabla_km.groupby('IdVehicule')['KmSaisie'].unique()
```

```python
def get_stats(group):
    return {'min': group.min(), 'max': group.max()}
```

```python
df=tabla_km.groupby('IdVehicule')['KmSaisie'].apply(get_stats).unstack()
```

```python
df['dif_km']=df['max']-df['min']
```

```python
df.head()
```

```python
tabla_consumo=consumos.loc[:,['DateHeure','IdVehicule','Quantite']]
tabla_consumo.head()
```

```python
tabla_consumo.sort_values('DateHeure').groupby('IdVehicule').last()
```

```python
df.head()
```

```python
Ult_Rep=tabla_consumo.sort_values('DateHeure').groupby('IdVehicule').last()
```

```python
df['Ult_Rep']=Ult_Rep

df.head()
```

```python
df['Quantite']=tabla_consumo.groupby('IdVehicule')['Quantite'].apply(sum)
```

```python
df['Consumo_Real']=df['Quantite']-df['Ult_Rep']
df.head()
```

```python
df['Estimación']=(df['Consumo_Real']/df['dif_km']*100).where((df['dif_km']>0)&(df['dif_km']<400000),-1)
```
