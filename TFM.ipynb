Puesto que como vemos las variables tienen diferentes escalas,procedemos a su normalización mediante Sklearn Normalizer

```python
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
```

Selección de variables para modelos posteriores

```python
# Separación de la variable objetivo y las explicativas en el dataframe sin Dia_semana
target = 'Usos_bicis_total'
features = list(scaled_bicimad.columns)
features.remove('Usos_bicis_total')
features.remove('Dia')
features.remove('Dia_semana')

x = scaled_bicimad[features]
y = scaled_bicimad[target]
```

```python
# Separación de la variable objetivo y las explicativas en el dataframe con Dia_semana con dummies
target_d = 'Usos_bicis_total'
features_d = list(scaled_bicimad_dummies.columns)
features_d.remove('Usos_bicis_total')
features_d.remove('Dia')

x_d = scaled_bicimad_dummies[features_d]
y_d = scaled_bicimad_dummies[target_d]
```

```python
# Selección de Variables mediante Variance Threshold sin variable Dia_semana

from sklearn.feature_selection import VarianceThreshold

var_th = VarianceThreshold(threshold = 0.05)
x_var = var_th.fit_transform(x)

print("Variables originales ", x.shape[1])
print("Variables finales ", x_var.shape[1])

print("Listado de variables ", np.asarray(list(x))[var_th.get_support()])

# Como vemos ninguna variable tiene varianza suficiente para ser considerada importante de cara a explicar 
# la variabilidad 
```

```python
# Selección de Variables mediante Variance Threshold con variable Dia_semana

var_d_th = VarianceThreshold(threshold = 0.05)
x_d_var = var_d_th.fit_transform(x)

print("Variables originales ", x_d.shape[1])
print("Variables finales ", x_d_var.shape[1])

print("Listado de variables ", np.asarray(list(x_d))[var_d_th.get_support()])

# Como vemos ninguna variable tiene varianza suficiente para ser considerada importante de cara a explicar 
# la variabilidad 
```

```python
# Selección de Variables mediante Select K-Best sin variable Dia_semana( con k=12 )

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression 
from sklearn.feature_selection import chi2 

var_sk = SelectKBest(f_regression, k = 12)
x_sk = var_sk.fit_transform(x, y)

print("Variables finales ", x_sk.shape[1])

print("Listado de variables ", np.asarray(list(x))[var_sk.get_support()])
```

Selección de Variables mediante VIF

```python
from sklearn.linear_model import LinearRegression

def calculateVIF(data):
    features = list(data.columns)
    num_features = len(features)
    
    model = LinearRegression()
    
    result = pd.DataFrame(index = ['VIF'], columns = features)
    result = result.fillna(0)
    
    for ite in range(num_features):
        x_features = features[:]
        y_feature = features[ite]
        x_features.remove(y_feature)
        
        x = data[x_features]
        y = data[y_feature]
        
        model.fit(data[x_features], data[y_feature])
        
        result[y_feature] = 1/(1 - model.score(data[x_features], data[y_feature]))
    
    return result
```

```python
def selectDataUsingVIF(data, max_VIF = 5):
    result = data.copy(deep = True)
    
    VIF = calculateVIF(result)
    
    while VIF.as_matrix().max() > max_VIF:
        col_max = np.where(VIF == VIF.as_matrix().max())[1][0]
        features = list(result.columns)
        features.remove(features[col_max])
        result = result[features]
        
        VIF = calculateVIF(result)
        
    return result
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import math

# Modelo para realizar los ajustes
model = LinearRegression()

# Variable para almecena los índices de la lista de atributos usados
feature_order =  []
feature_error = []

# Iteración sobre todas las variables
for i in range(len(features_d)):
    idx_try = [val for val in range(len(features_d)) if val not in feature_order]
    iter_error = []

    for i_try in idx_try:
        useRow = feature_order[:]
        useRow.append(i_try)

        use = x_d[x_d.columns[useRow]]

        model.fit(use, y_d)
        rmsError = np.linalg.norm((y_d - model.predict(use)), 2)/math.sqrt(len(y_d))
        iter_error.append(rmsError)

    pos_best = np.argmin(iter_error)
    feature_order.append(idx_try[pos_best])
    feature_error.append(iter_error[pos_best])

for i in range(len(features_d)):
    print("En el paso", i, "se ha insertado la variable", 
        features_d[feature_order[i]], "con un error", feature_error[i])
```
